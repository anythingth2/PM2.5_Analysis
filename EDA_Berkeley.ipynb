# EDA Berkeley dataset

This part will explore insight from berkeley's PM2.5 dataset (http://berkeleyearth.org/).

The dataset was scraping from website by using modified code from https://github.com/worasom/aqi_thailand.

The dataset includes PM2.5 value, location of sensor, city name and etcs.

```python
import pandas as pd 
```

```python
region_df = pd.read_csv('dataset/berkeleyearth/region.csv')
region_df.rename({'Unnamed: 0': 'id'}, axis=1, inplace=True)
sensor_df = pd.read_csv('dataset/berkeleyearth/sensor.csv')
sensor_df.rename({'Unnamed: 0': 'id'}, axis=1, inplace=True)
sensor_df['timestamp'] = pd.to_datetime(sensor_df['timestamp'])
sensor_df['year'] = sensor_df['timestamp'].apply(lambda timestamp: timestamp.year)
sensor_df['month'] = sensor_df['timestamp'].apply(lambda timestamp: timestamp.month)
sensor_df['day'] = sensor_df['timestamp'].apply(lambda timestamp: timestamp.day)
```

# PM2.5 monthly Trend

Visualize monthly trend of PM2.5 data by using color from Red(Jan) to Black(Dec).
This part aims to explore seasonal trend of pm2.5.

```python
sample_city = region_df['city'].sample(10).sort_values()
print(sample_city)
sample_sensor = sensor_df[sensor_df['city'].isin(sample_city)]
```

```python
colors = ['red', 'pink', 'orange', 'yellow', 'green', 'cyan',
          'blue', 'indigo', 'purple', 'brown', 'grey', 'black']
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
ax = None


def plot_city(city):
    name = city.name
    city = city.copy()

    city['month'] = city['timestamp'].apply(lambda timestamp: timestamp.month)
    global ax
    ax = None

    def plot_month(monthly_city):
        month = monthly_city['month'].iloc[0]
        global ax
        ax = monthly_city.plot(x='timestamp', y='pm', title=name, label=months[(month-1) % len(months)],
                               style='.', color=colors[(month-1) % len(colors)], ax=ax, figsize=(24, 8))
    city.groupby('month').apply(plot_month)
    return ax
```

```python
sample_sensor.groupby('city').apply(plot_city)
```

Obviously, winter season (november to february) has pm2.5 more than other seasons.

# Seasonal trending

To prove seasonal trend, we visualize mean of PM2.5 value of the month in same figure.

```python
sample_city = sensor_df.groupby('city')['pm'].mean().sort_values(ascending=False).index[:10].tolist()
```

```python
# sample_city = region_df['city'].sample(10).sort_values()
print(sample_city)
sample_sensor = sensor_df[sensor_df['city'].isin(sample_city)]
```

```python
colors = ['red',  'green',
          'blue', 'brown', 'grey', 'black']


def plot_monthly_trending(city):
    city_name = city.name
    city = city.copy()
    city['month'] = city['timestamp'].apply(lambda timestamp: timestamp.month)
    city['year'] = city['timestamp'].apply(lambda timestamp: timestamp.year)
    city['color'] = city['year'].apply(lambda year: colors[year % len(colors)])

    monthly_city = city.groupby(['year', 'month']).mean()
    ax = None
    for year, group in monthly_city.reset_index().groupby('year'):
        color_name = colors[year % len(colors)]
        ax = group.plot(ax=ax, x='month', y='pm', kind='line',
                        color=color_name, label=str(year), title=city_name)
    return ax


sample_sensor.groupby('city').apply(plot_monthly_trending)
```

# Daily Max pm2.5 data

```python
max_daily_sensor = sensor_df.groupby(['city', 'year', 'month', 'day']).apply(
    lambda daily_sensor: sensor_df.iloc[daily_sensor['pm'].idxmax()])
```

```python
max_daily_sensor.head(10)
```

```python
sample_city = region_df['city'].sample(10).sort_values()
print(sample_city)
sample_sensor = max_daily_sensor[max_daily_sensor['city'].isin(sample_city)].set_index('id')
```

```python
sample_sensor.head()
```

```python
sample_sensor.groupby('city').apply(lambda city_sensor: city_sensor.plot(
    x='timestamp', y='pm', kind='line', title=city_sensor.name, figsize=(24, 8)))
```

# Interpolation

```python
target_city = 'ubonratchathani'
city_sensor = sensor_df[sensor_df['city'] == target_city]
```

## Before Interpolation

```python
city_sensor.plot(x='timestamp', y='pm', style='.', figsize=(24, 8))
```

## After Interpolation

```python
interpolated_city_sensor = city_sensor.set_index('timestamp').resample('H').mean()
interpolated_city_sensor['pm'] = interpolated_city_sensor['pm'].interpolate()
interpolated_city_sensor.reset_index(inplace=True)
```

```python
interpolated_city_sensor.plot(x='timestamp', y='pm', style='.', figsize=(24, 8))
```
